```python
import pandas as pd
```

```python
reading_df = pd.read_csv("../data/raw/reading_answers.csv")
listening_df = pd.read_csv("../data/raw/listening_answers.csv")
writing_df = pd.read_csv("../data/raw/writing_samples.csv")
```

```python
reading_df.head()
```

```python
listening_df.head()
```

```python
writing_df.head()
```

```python
reading_df.isnull().sum()
listening_df.isnull().sum()
writing_df.isnull().sum()
```

```python
writing_df = writing_df.dropna(subset=["overall_band"])
writing_df.isnull().sum()
```

```python
reading_df["text"] = reading_df["answer_text"]
listening_df["text"] = listening_df["answer_text"]
writing_df["text"] = writing_df["writing_sample"]
```

```python
reading_df["label"] = reading_df["correct_label"]
listening_df["label"] = listening_df["correct_label"]
writing_df["label"] = writing_df["overall_band"]
```

```python
reading_df["task"] = "reading"
listening_df["task"] = "listening"
writing_df["task"] = "writing"
```

```python
reading_clean = reading_df[["text", "label", "task"]]
listening_clean = listening_df[["text", "label", "task"]]
writing_clean = writing_df[["text", "label", "task"]]
```

```python
full_df = pd.concat([reading_clean, listening_clean, writing_clean], ignore_index=True)
full_df.head()
```

```python
full_df.to_csv("../data/processed/oet_samples_small.csv", index=False)
```
